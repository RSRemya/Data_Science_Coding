# Encapsulation

Its used to restrict access to methods and variables. In encapsulation, code and data are wrapped together within a single unit from being modified by accident.

```python
# single underscore before a variable means that the variable is protected, can be access them inside class and subclass
# double underscore before a variable means that the variable is private, only within the class its accesible
```

```python
class Hello:
    def __init__(self,name):
        self.a = 10
        self._b = 20
        self.__c = 30
        

        
hello = Hello('name')
print(hello.a)
print(hello._b)
print(hello.__c)
```

```python
#Since c is private variable .... its not accessible outside the class Hello
# Unlike other programming languages, python does not have keyword like private or protected, thats why _ and __ is used
```

```python
class Car:
    def __init__(self, speed, color):
        self.speed = 200
        self.color = "red"

ford = Car(200, 'red')
honda = Car(250, 'blue')
audi = Car(300, 'black')

ford.speed="sjdfosdj"
# this will be the final value of speed, but such insane alteration of values of variable should be avoided
# by using encapsulation

print(ford.speed)
print(ford.color)
```

```python
# To encapsulate the code some more functions are required like - set_speed, get_speed & set_color, get_color

class Car:
    def __init__(self, speed, color):
        self.speed = speed
        self.color = color

    def set_speed(self, value):
        self.speed = value

    def get_speed(self):
        return self.speed


    def set_color(self, value):
        self.color = value

    def get_color(self):
        return self.color

ford = Car(200, 'red')
honda = Car(250, 'blue')
audi = Car(300, 'black')

ford.set_speed(300)

print(ford.get_speed())
print(ford.get_color())
```

```python
# But if later its altered as follows:
ford.speed = "hdcdkjkofd"
print(ford.get_speed())
```

```python
# The variable value is getting altered, so make the variable speed and color as private variable

class Car:
    def __init__(self, speed, color):
        self.__speed = speed
        self.__color = color

    def set_speed(self, value):
        self.__speed = value

    def get_speed(self):
        return self.__speed


    def set_color(self, value):
        self.__color = value

    def get_color(self):
        return self.__color

ford = Car(200, 'red')
honda = Car(250, 'blue')
audi = Car(300, 'black')

ford.set_speed(300)
ford.speed="hdcdkjkofd"

print(ford.get_speed())
print(ford.get_color())
```
